13. Armstrong Number

```python
n = int(input("Enter the number : "))
temp = n
length = 0
total = 0
while temp > 0 :
  length += 1
  temp //= 10

temp = n

while temp > 0 :
  last_digit = temp % 10
  total = total + (last_digit ** length)
  temp //= 10

if total == n :
  print("-----------------")
  print(f"""number | total\n{n}    | {total}\n-----------------\nArmstrong number""")
else :
  print("not armstrong")
```

20. Count Occurrences of every one Digit

23. Check Duck Number

A Duck Number is a number that:
*   Contains at least one 0.
*   The 0 is not at the beginning of the number.

24. Convert Decimal to Binary
* it is not actually stored as binary in memory. Python stores it as the decimal integer one thousand one hundred one (1101).

* The true binary representation of 13 inside the computer is different. We're just displaying the binary digits as an integer for convenience.

* That's why many interviewers accept the string solution as the more correct representation of a binary number.

```python
num = int(input("Enter the number :"))

if num == 0 :
  print("Binary Number = 0")
else :
  binary_number = ""

while num > 0 :
  remainder = num % 2
  binary_number = str(remainder) + binary_number
  num //= 2

print("Binary Number =",binary_number)
```

26. Convert Decimal to Hexadecimal (Logic Only)


*   Divide the number by 16.
| Decimal | Hex |
| ------- | --- |
| 0–9     | 0–9 |
| 10      | A   |
| 11      | B   |
| 12      | C   |
| 13      | D   |
| 14      | E   |
| 15      | F   |




```python
num = int(input("Enter the number : "))
hexa = ""

while num > 0 :
  remainder = num % 16
  if remainder < 10 :
    hexa = str(remainder) + hexa
  elif remainder == 10 :
    hexa = "A" + hexa
  elif remainder == 11 :
    hexa = "B" + hexa
  elif remainder == 12 :
    hexa = "C" + hexa
  elif remainder == 13 :
    hexa = "D" + hexa
  elif remainder == 14 :
    hexa = "E" + hexa
  elif remainder == 15 :
    hexa = "F" + hexa

  num //= 16

print("HEXADECIMAL NUMBER : ", hexa)

```

27. Find GCD (HCF) of Two Numbers
* GCD (Greatest Common Divisor) or HCF (Highest Common Factor) is the largest number that divides both numbers exactly.

```python
num1 = int(input("Enter the first number : "))
num2 = int(input("Enter the  second number : "))

i = 1
gcd = 1

while i <= num1 and i <= num2 :
  if num1 % i  == 0 and num2 % i == 0 :
    gcd = i
  i += 1

print("GCD =",gcd)
```

28. Find LCM of Two Numbers (Least common Multiple)

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

lcm = max(num1, num2)

while True:
    if lcm % num1 == 0 and lcm % num2 == 0:
        print("LCM =", lcm)
        break

    lcm += 1
```

29. Check Strong Number
* A Strong Number is a number whose sum of the factorials of its digits is equal to the number itself.

```python
num = int(input("Enter the number :")) # 145
temp = num
total  = 0

while temp > 0 :
  digit = temp % 10
  fact = 1
  i = 1
  while i <= digit :
    fact = fact * i
    i += 1
  temp //= 10
  total = total + fact

if num == total :
  print("strong number")
else :
  print("not strong number")


```

30. Check Neon Number

31. Find All Armstrong Numbers Between 1 and 1000

```python
num = 1

while num <= 1000 :
  temp = num
  no_of_digit = 0

  while temp > 0 :
    no_of_digit += 1
    temp //= 10

  temp = num
  total = 0

  while temp > 0 :
    digit = temp % 10
    total = total + digit ** no_of_digit
    temp //= 10

  if num == total :
    print(num, end = " ")

  num += 1
```

33.Check Circular Prime - A Circular Prime is a prime number such that all rotations of its digits are also prime.

** Example:
* 197

** Rotations:
* 197
* 971
* 719

```python
num = input("Enter a number: ")

def is_prime(n):
    if n < 2:
        return False

    i = 2
    while i * i <= n:
        if n % i == 0:
            return False
        i += 1

    return True

rotation = num
count = len(num)

is_circular = True

while count > 0:

    if not is_prime(int(rotation)):
        is_circular = False
        break

    rotation = rotation[1:] + rotation[0]
    print(rotation)
    count -= 1

if is_circular:
    print("Circular Prime")
else:
    print("Not Circular Prime")
```
