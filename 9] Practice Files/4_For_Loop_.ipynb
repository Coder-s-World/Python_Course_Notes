```python
# 1.Prime Numbers in a Range. Ex 1 - 100
# Prime number = The numbers which divide by either number itself or 1.


start = int(input("Enter start: "))
end = int(input("Enter end: "))

count = 0

for num in range(start, end + 1):
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            print(num, end=" ")
            count += 1

print("\nTotal prime numbers:", count)
```

```python
# 2. Armstrong Number
# Armstrong Number - An Armstrong number is a number that is equal to the sum of its digits raised to the power of the
# number of digits.

num = int(input("Enter the number :- "))

length = len(str(num))
total = 0

for digit in str(num) :
  total +=  int(digit) ** length

if total == num :
  print("armstrong number")
else :
  print("not armstrong number")
```

```python
# 3. Perfect Number
# Perfect Number - A Perfect Number is a number that is equal to the sum of its proper divisors (excluding itself).

# Examples
# 6 → 1 + 2 + 3 = 6 ✔
# 28 → 1 + 2 + 4 + 7 + 14 = 28 ✔

total = 0
totalDiv = 0
div = []

num = int(input("enter the number :- "))

for i in range(1,num) :
  if num % i == 0 :
    div.append(str(i))
    total += i
print("Divisors addition : ", " + " .join(div),"=",total)

if total ==  num :
   print("perfect number")
else :
   print("not perfect number")
```

```python
# 4.Factorial of Number
# Factorial of a number is the product of all positive integers from 1 to that number.

# For example:
# 5! = 1 × 2 × 3 × 4 × 5 = 120
# So, 5! = 120.

num = int(input("Enter the number :- "))
total = 1
iter = []
for i in range(1,num+1) :
  total =  total * i
  iter.append(str(i))


print("factorial of number : "," + ".join(iter)," = ",total)
```

```python
# 5. Strong Number
# Strong Number - A Strong Number is a number whose sum of the factorials of its digits equals the number itself.

# Example
# 145 = 1! + 4! + 5!
# 145 = 1 + 24 + 120 = 145 ✔

num = int(input("enter the number :- "))
total = 0

for digit in str(num):
  fact = 1
  for j in range(1,int(digit)+1) :
    fact = fact * j
  total = total + fact

print("factorial of no : ", total)

if total == num :
  print("string number")
else :
  print("not strong number")
```

```python
# 7. Pascal's Triangle
# Print Pascal's Triangle for n rows.

# Example:

#       1
#      1 1
#     1 2 1
#    1 3 3 1

# Every row starts and ends with 1, and each middle value is the sum of the two values above it.

n = int(input("Enter number of rows: "))

for i in range(n):
    num = 1

    # Print spaces
    for j in range(n - i - 1):
        print(" ", end=" ")

    # Print numbers
    for j in range(i + 1):
        print(num, end="   ")
        num = num * (i - j) // (j + 1)

    print()
```

```python
# 8. Number Pyramid
# 1
# 121
# 12321
# 1234321
# 123454321

n = 5

for i in range(1, n + 1):

    # Increasing numbers
    for j in range(1, i + 1):
        print(j, end="")

    # Decreasing numbers
    for j in range(i - 1, 0, -1):
        print(j, end="")

    print()
```

```python
# 9. Hollow Square Pattern
# *****
# *   *
# *   *
# *   *
# *****

n = int(input("Enter size: "))

for i in range(n):
    for j in range(n):

        if i == 0 or i == n - 1 or j == 0 or j == n - 1:
            print("*", end="")
        else:
            print(" ", end="")

    print()
```

```python
# 10. Diamond Pattern
#     *
#    ***
#   *****
#  *******
# *********
#  *******
#   *****
#    ***
#     *

n = 5

# Upper half
for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2 * i - 1))

# Lower half
for i in range(n - 1, 0, -1):
    print(" " * (n - i) + "*" * (2 * i - 1))
```

```python
# 11. Count Character Frequency

# Count the frequency of each character in a string without using built-in counting functions.

# Example:

# Input: banana
# Output:
# b = 1
# a = 3
# n = 2

text = input("enter the text : ")
visited = " "
for ch in text :
  if ch not in visited :
    count = 0
    for c in text :
      if ch == c :
        count += 1
    print(f"{ch} = {count}")
    visited += ch


```

```python
# 12. Remove Duplicate Characters
# Remove duplicate characters from a string using only loops.

# Example:

# Input : programming
# Output : progamin

text = input("Enter the text : ")

seen = set()
result = " "

for ch in text :
  if ch not in seen :
    seen.add(ch)
    result += ch
print(result)
```

```python
# 13. Find Second Largest Number
# Find the second largest number in a list without sorting.

# Uses of map() in Python
  # map() applies a function to every element of an iterable (list, tuple, etc.).
  # It helps avoid writing explicit loops for simple transformations.
  # Commonly used to convert data types (str → int, str → float).
  # Useful for performing the same operation on all elements.
  # syntax- map(function,iterable) Function=What operation should be applied? ,Iterable=On which collection should it be applied?


# The reason many programmers use float('-inf') is that it works correctly even when the list contains only negative numbers, such as:
# [-50, -20, -10, -30]
# Negative infinity is a value that is smaller than every possible number.

list1 = list(map(int,input("enter the  element for list : ").split()))
print("list is : ",list1)

largest = second_largest = float('-inf')

for num in list1 :
  if num > largest :
    largest = num
print("largest number : " ,largest)

for num in list1 :
  if num != largest and num > second_largest :
    second_largest = num
print("second largest : ",second_largest)
```

```python
# 14. Matrix Addition
# Add two matrices using nested for loops.

A = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

B = [
    [4, 5, 6],
    [7, 8, 9],
    [1, 2, 3]
]

result = []

for i in range(len(A)):
    row = []

    for j in range(len(A[0])):
        row.append(A[i][j] + B[i][j])

    result.append(row)

for row in result:
    print(row)
```

```python
# 15. Matrix Multiplication

# Multiply two matrices using nested for loops.

A = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

B = [
    [4, 5, 6],
    [7, 8, 9],
    [1, 2, 3]
]

result = []

for i in range(len(A)):
    row = []

    for j in range(len(A[0])):
        row.append(A[i][j] * B[i][j])

    result.append(row)

for row in result:
    print(row)

```

```python
# 16. Transpose of Matrix
# Find the transpose of a matrix using nested loops.

# What is a Transpose?
# Transpose means converting rows into columns and columns into rows.

# Example

# Original Matrix:

# 1 2 3
# 4 5 6
# 7 8 9

# Transpose:

# 1 4 7
# 2 5 8
# 3 6 9

# Notice:

# A[0][1] = 2  →  Transpose[1][0] = 2
# A[2][1] = 8  →  Transpose[1][2] = 8

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

transpose = []

for j in range(len(matrix[0])):

    row = []

    for i in range(len(matrix)):
        row.append(matrix[i][j])

    transpose.append(row)

for row in transpose:
    print(row)
```

```python
# 17. Multiplication Table Grid

# Print:

# 1   2   3   4   5
# 2   4   6   8  10
# 3   6   9  12  15
# 4   8  12  16  20
# 5  10  15  20  25
# \t stands for a tab character (horizontal tab). - It creates a larger space between outputs, making columns look aligned.

n = 5

for i in range(1, n + 1):
    for j in range(1, n + 1):
        print(i * j, end="\t")
    print()
```

```python
# 18. Find Missing Number
# Given:
# [1, 2, 3, 5, 6]

# Find the missing number using loops.

list1 = [1,2,3,4,5,10,11,13,6,7,9]
largest  = max(list1)

for i in range(1,largest+1) :
  found = False
  for num in list1 :
    if num == i :
      found = True
      break

  if not  found   :
    print("missing number : ", i)
```

```python
# 19. Anagram Checker
# Check whether two strings are anagrams using loops.

# Example:

# listen
# silent

# Output:
# Anagram

str1 = input("enter the string 1 :")
str2 = input("enter the string 2 :")

if len(str1) != len(str2) :
  print("not anagram")

else :
  is_anagram = True

  for ch in str1 :
    count_str1 = 0
    count_str2 = 0

    for c in str1 :
      if c == ch :
        count_str1 += 1
    for c in str2 :
      if c == ch :
        count_str2 += 1

    if count_str1 != count_str2 :
      is_anagram = False
      break

  if is_anagram :
    print("is anagaram")
  else :
    print("not anagaram")


```

```python
# 20. Longest Word Finder
# Find the longest word in a sentence using a for loop.

# Example:

# Input:
# Python is an amazing programming language

# Output:
# programming

sentence = input("Enter a sentence: ")

words = sentence.split()

longest = ""

for word in words:
    if len(word) > len(longest):
        longest = word

print("Longest word:", longest)
```
