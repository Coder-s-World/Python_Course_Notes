2. If-Else Statement

```python
# Basic

# 1) Check whether a number is even or odd.
number = int(input("Enter the number : "))

if num % 2 == 0 :
  print(f"{number} is  even number")
else :
  print(f"{number} is odd number")


# 2) Check whether a person is minor or adult.
Age = int(input("Enter the age : "))

if Age > 18 :
  print("person is adult")
else :
  print("person is minor")


# 3) Find the larger of two numbers.
num1 = int(input("Enter number 1 - "))
num2 = int(input("Enter number 2 - "))

if num1 > num2 :
  print(f"{num1} is greater than {num2}")
else :
  print(f"{num2} is greater than {num1}")


# 4) Check whether a number is positive or negative.
num = int(input("Enter the number - "))

if num  > 0 :
  print("number is positive")
else :
  print("number is negative")


# 5) Check whether a character is a vowel or consonant.
vowels = "AEIOUaeiou"
char = input("Enter the character - ")

if char in vowels :
  print("vowel")
else :
  print("consonent")
```

```python
# Intermediate

# 1) Check whether a year is a leap year.
year = int(input("Enter the year - "))
if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0 :
  print("leap year")
else :
  print("not leap year")

# 2) Check if a student passed or failed.
percentage = int(input("Enter the percentage : "))
if percentage >= 35 :
  print("passed")
else :
  print("failed")


# 3) Check whether a number is divisible by 2 or not.
num = int(input("Enter the number - "))
if num % 2 == 0 :
  print("number is divisible by 2")
else :
  print("number is not divisible by 2")


# 4) Compare two strings and print if they are equal.
text1 = input("Enter the text 1 : ")
text2 = input("Enter the text 2 : ")
if text1 == text2 :
  print("both the strings are equal")
else :
  print("both strings are not equal")


# 5) Check whether a temperature is above or below freezing.
temp = float(input("Enter the temperature - "))
if temp > 0 :
  print("above freezing point")
elif temp == 0 :
  print("at freezing point")
else :
  print("below freezing point")
```

```python
# Advanced

# 1) Check whether a string is empty.
text = input("Enter the text - ")
if len(text) < 1 :
  print("empty string")
else :
  print(f"string has {len(text)} characters")

# 2) Determine if a password is strong or weak.
password = input("Enter the password - ")

has_lower = False
has_Upper = False
has_digit = False

for ch in password :
  if ch.islower() :
    has_lower = True
  elif ch.upper() :
    has_upper = True
  elif ch.isdigit():
    has_digit = True

if len(password) >= 8 and has_lower and has_upper and has_digit :
  print("strong password")
else :
  print("weak password")
```

```python
# Basic
# Check if a number is positive, then check if it is even.
num = int(input("Enter the number - "))
if num > 0 :
  print("positive")
  if num % 2 == 0 :
    print("even")

# Check if a student passed, then check if the grade is A.
marks = int(input("Enter the marks : "))
if marks >= 35 :
  print("passed")
  if marks >= 65 :
    print("Grade A")

# Check if age is above 18, then check if driving license exists.
age  = int(input("Enter the age : "))
license =  input("Do you have a driving license (yes/no) : ")

if age > 18 :
  if license == "yes" :
    print("eligible for driving")
  else :
    print("driving license required")
else :
    print("age criteria does not match for driving license")


# Check if salary is above ₹20,000, then check if eligible for a loan.
salary = int(input("Enter the salary : "))
cibil_score = int(input("Enter the cibil score : "))

if salary > 20000 :
  if cibil_score >= 750 :
    print("loan approved")
  else :
    print("loan does not approved due to low cibil score")
else :
  print("salary criteria not matched")


# Check if a number is divisible by 2, then by 4.
num = int(input("Enter the number : "))
if num % 2 == 0 :
  print("number is divisible by 2")
  if num % 4 == 0 :
    print("number is also divisible by 4")
```

```python
# Intermediate
# Find the largest among three numbers using nested if.

# Input three numbers
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

# Nested if logic to find the largest number
if num1 >= num2:
    if num1 >= num3:
        largest = num1
    else:
        largest = num3
else:
    if num2 >= num3:
        largest = num2
    else:
        largest = num3

print(f"The largest number is: {largest}")


# Check username first, then password.
username = input("Enter username : ")
password = input("Enter password : ")

if username == "Mayu_44" :
  print("correct username")
  if password == "123@Python" :
    print("correct password")
  else :
    print("incorrect password")
else :
  print("incorrect username")

# Check if a year is leap year and also century year.

# Check if marks are above 40, then determine distinction.

# Check if an employee is permanent, then calculate bonus.
```

6. Short-Hand If-Else Statement (Ternary Operator)

```python
# Basic
# Print "Even" or "Odd".
num = int(input("enter the number : "))
print("even") if num % 2 == 0 else print ("odd")

# Print "Pass" or "Fail".
marks = int(input("enter the number : "))
print("pass") if marks >= 35 else print("fail")

# Find larger of two numbers.
num1 = int(input("enter the first number : "))
num2 = int(input("enter the second number : "))
print("num2 is greater") if num2  > num1 else print("num1 is greater")

# Print "Adult" or "Minor".
age = int(input("enter the age : "))
print("adult") if age >= 18 else print("minor")

# Print "Positive" or "Negative".
num = int(input("enter the number : "))
print("positive") if num > 0 else print("negative")
```
